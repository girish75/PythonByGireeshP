```python
# 1. Function set
# suppose we have list with duplicate values and wanted to remove duplicates out of it
my_list = [1,2,3,4,5,5]
print(my_list)
#remove duplicate using set() function
print(set(my_list)) # you can use this for example to remove duplicate email address

# set object do not support indexing
#my_set[0] # set object do not support indexing so to access a value directly 
print(my_set)
print(1 in my_set)
print(len(my_set))
#convert set into list
print(list(my_set))
#copy () method
your_set =my_set.copy()
print(your_set)
#clear() method
my_set.clear()
print(my_set)

#methods from ven diagram
# .difference()
# .discard() and .remove
# .difference_update()
# .intersection()
# .isdisjoint()
# .union()
# .issubset()
# .issuperset()


# .difference()  it will not remove the values which coexist from your_set but return the values which are do not coexist from first set. 
my_set = {1,2,3,4,5}
your_set = {4,5,6,7,8,9}
print(my_set.difference(your_set)) 
print(my_set) # you can see that it Do not actually remove .. 
# .discard() - it will remove the item mentioned if exist in the set. Return None. In place discard\ remove. 
print(my_set.discard(4)) # this will return None
print(my_set)
print(my_set.remove(5))
print("my_set-----------", my_set) # Both discard and remove are same. 

# .difference_update()  it will actually remove the values which coexist from your_set. Return None. In place replacement. 
my_set = {1,2,3,4,5}
your_set = {4,5,6,7,8,9}
print(my_set.difference_update(your_set))
print(my_set)

# .intersection - only return values which coexist or duplicate in both the sets
# re assign values to these sets. 
my_set = {1,2,3,4,5}
your_set = {6,7,8,9,4,5}

print(my_set.intersection(your_set))
print(my_set & your_set) # this is short hand notation of intersection. 

#.isdisjoint() - True means if set have nothing in common. False means they have something in common.  
print(my_set.isdisjoint(your_set)) # will return False. 

# .union() - 
full_set = my_set.union(your_set)

print(my_set.union(your_set)) # create a new set with all the values without duplicate ones. 
print("Print full set")
print(full_set)

# OR 
print(my_set | your_set) # Shorthand for Union whih also create a new set with all the values without duplicate ones. 

#.issubset()
print(my_set.issubset(full_set))
#.issuperset()
print(full_set.issuperset(my_set))
```
