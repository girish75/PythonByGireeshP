```python
#List are mutable unlike strings which are not because in list we can change any of its index value. 
#List are one of data structure like tuple and dict. 

# List Declaration or creating a list - Way 1
amazon_cart = [
    'notebooks', 
    'sunglasses', 
    'toys',
    'grapes'
    ]

# List unpacking - assinging each item of list into different variables. 
print("List unpacking")
a,b,c,d = [
    'notebooks', 
    'sunglasses', 
    'toys',
    'grapes'
    ]
print(a)
print(b)
print(c)
print(d)

# List unpacking - assinging each item of list into different variables. 
print("List unpacking - assign few items in variables and let remaining in a list")

a,b,*other, last = [
    'notebooks', 
    'sunglasses', 
    'toys',
    'grapes',
    'protien',
    'Suppliments',
    'medicine'
    ]
print(a)
print(b)
print(other)
print(last)


# Lists are mutable - let us access and then change its value
print (f"First index value of list is : {amazon_cart[0]}")


#list sliciing 
print (f"List slicing - index value of list is : {amazon_cart[0:2]}")
#list is mutable
amazon_cart[0] = 'laptop' # this will change the cart forever 

print (f"Full cart value {amazon_cart}")
#Will this copy the list? -- No its not. 
amazon_cart1 = amazon_cart
amazon_cart1[0] = 'notebook' 
print ("this will impact both variables  amazon_cart1 and amazon_cart")
print(amazon_cart1)
print(amazon_cart)

#How to copy the entire cart to new cart ?  -- Trick

amazon_cart2 = amazon_cart[:]
amazon_cart2[0] = 'computer' 
print( "this will not impact both variables  amazon_cart2 and amazon_cart but cart2 is an another copy\ n")
print(amazon_cart2)
print(amazon_cart)

```

```python
#List methods
amazon_cart = [
    'notebooks', 
    'sunglasses', 
    'toys',
    'grapes'
    ]
# 1. append method - This method insert value to the end of the list. This method does not return anything means it returns None. 
amazon_cart.append(100)
print(amazon_cart)

# No use of assigning the return value of append method of list. 
cart = amazon_cart.append(200)

print(amazon_cart)
print(cart)

# 2. insert method - Same as append but at desired index value it add the new value in list.  This method does not return anything means it returns None.
print("Inserting value at 3rd index value of the cart")
amazon_cart.insert(2,100)
print(amazon_cart)
# "3. extend method - To append multiple items into an existing list - note: use listed or tuple values"
print("3. extend method - To append multiple items into an existing list - note: use listed or tuple values")

e = amazon_cart.extend([100,101, 102])
print(amazon_cart)
print("No return value")
print(e)

# There are 3 removing methods as pop method, remove and clear method
# ====================================================================
# 4. pop mentod - This menthod returns the value it removes. Default removal is from end of the list. 
a = amazon_cart.pop() # remember the cart value before execution of pop was  ['notebooks', 'sunglasses', 100, 'toys', 'grapes', 100, 200]
print (amazon_cart)
print(a)
# pop menthod with index to remove the value from specified index 
a = amazon_cart.pop(1) # remember the cart value before execution of pop was  ['notebooks', 'sunglasses', 100, 'toys', 'grapes', 100, 200]
print (amazon_cart)
print("Pop has return value")
print(a)

# 5 remove method - This mentod require the actual Value to be removed and not the index. note Value should exist else it will through an error
#amazon_cart.remove() # error
ret = amazon_cart.remove(100) # It will remove first occurance of value)
print(amazon_cart)
print("No return value of the remove method")
print(ret)

# 6 clear method - completely removes all the items

ret = amazon_cart.clear()

print(amazon_cart)

print("No return value of the clear method")
print(ret)


# 7. Index (value, optional - startindexlocation, optional -stopindexlocation) method - if the value does not exist in the list then it will return an Error (not good)
# ====================================================================
amazon_cart = [
    100, 
    200, 
    100,
    'a',
    'b',
    'c'
]
print(f"Index value of 200 is {amazon_cart.index(200)}")
print(f"Index value of 100 is {amazon_cart.index(100)}")
# print(f"Index value of 400 is {amazon_cart.index(400)}") # Produce an Error

# 8. Keyword for list -- How to know if the value exist in a list
#=============================================================================

# in keyword

print(400 in amazon_cart)
print(100 in amazon_cart)

# 9. Count method  - how many times an item or a value exist in the list. It return the counted value
# ==============================
a = amazon_cart.count(100)
print(a)

# 10. Sort method - In place method - Do not return anything but sort the items within the list.
# Sort only sort single type of values in the list. Means list should not have int and str or multiple data types
print (amazon_cart) # orignial values of list
# print(amazon_cart.sort()) # Return None 
#amazon_cart.sort() # this will return error because amazon_cart has Str and Int type of values in it. 
amazon_cart1.clear()
amazon_cart1 = [ 'x', 'b', 'd', 'c', 'a']
amazon_cart1.sort()
print (amazon_cart1)

# OR
amazon_cart2 = [3,5,2,1]
amazon_cart2.sort()
print(amazon_cart2)

#10. Let us use built in function instead of method 
amazon_cart3 = ['x', 'w', 'd', 'c', 'a']
print(sorted(amazon_cart3)) # Sorted function returns sorted array value or create new copy . Do not perform In place as sort method. 
print(amazon_cart3) # still the orignial cart order values will remain same - not sorted. 

#11. How to copy the entire cart to new cart ?  -- Trick

# way 1 -- amazon_cart2 = amazon_cart[:]
# way 2 --
amazon_cart2 = amazon_cart.copy()
print(amazon_cart2)

# 12. Reverse method for the entire list
amazon_cart1 = ['x', 'w', 'd', 'c', 'a']
print(amazon_cart1)
amazon_cart1.sort()
print(amazon_cart1)
amazon_cart1.reverse()
print(amazon_cart1) # completely reverse the values of the list - all index
# another way to reverse a list. This is same as string. We use slicing. 
str = "Girish"  # for a string and the same applies to list
print(str[::-1])
print(amazon_cart1[::-1])

# 13. Range - to create a list 
print(range(1,100,2))

print(list(range(1,100,2)))

print(f"\n\t {list(range(101))}")

# 14. Joining a string and a list. 

print('!'.join(["hi", "my", "name", "is", "JOJO"]))
print(str.join(["hi", "my", "name", "is", "JOJO"]))
```
