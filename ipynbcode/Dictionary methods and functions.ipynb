```python
# -*- coding: utf-8 -*-
"""
Dictionaries: 
Dictionaries are perhaps the most flexible built in data type in Python. 
In other languages these are called map, hash table or object. 
It is also a data structure. 

The chief distinction is that in dictionaries, items are stored and fetched instead of positional offsets. 
list are ordered collection of objects

Disctionar are unordered collection of objects.
Json like Dictionaries. 

Properties of Dictionaries: 
>>Accessed by Keys and not by offset
>> Unordered collection of arbitary values distributed all across in memory so its values can be unordered when printed---Note it
>> Variable length, hetrogenious, arbitarily nestable.
>> Mutable mapping ---Note it
>> Table of object references (hash tables)
>> Key of dictionary has to be Unique else it will get replaced by its next occurrance
>> Key can be any immutable data type value as string, boolean, int etc. 

>> slicing and concatenation on Dictionaries are not applicable as you can fetch values only by keys and not by positions.
"""

Dic_structure = {'a' : 1, 'b' : 2 }
print(Dic_structure['b'])

#dict.keys() method will provide all the keys
Dic_structure = {'python' : 1, "java" : 2}
print(f"Print dictionary all the keys : {Dic_structure.keys()}" )

# adding key and values in a dictionary. 
print("adding key and values in a dictionary. ")
Dic_structure["python"] = 1.1
Dic_structure["Ruby"] = 0
Dic_structure[0] = "Hadoop"

# 1.keys() method to print all the keys of dictionary
print(Dic_structure.keys())
print(Dic_structure)

# 1. Trick -- printing data type of keys of the list
print("\n Printing data type of keys of the list")

[print(type(i)) for i in Dic_structure]

# 2. dict.values() method will provide all the values
print("\n dict.values() will provide all the values") 
print(" dict.values ", Dic_structure.values())

# 3. dict.items() - will print items of dictionary
print("\n dict.items() - will print items of dictionary")

print("\n Dictionary items are  ", Dic_structure.items())

# 4. dict.get() will provide one value only based on key entered. 
print("\n Dic_structure.get(python) method",  Dic_structure.get("python"))

# What if a key does not exist in dictionary using Dic_structure.get() method? - It will through None. 
# So one way to avoid that is to pass a default value. 

print("\n Dic_structure.get(ython) method ", Dic_structure.get("ython")) # will return None

print("\n Dic_structure.get(ython) method ", Dic_structure.get("ython", 0)) # will return 0

# 5. .dict() method can be used to create dictionary on fly 
user = dict(name = "JOJO")
print(user)

#2. Trick use keyword in
print("\n Use keyword in for a dictionary")
print('ython' in Dic_structure)

#6. dict.update() -- This a merge method. In case if key does not exist then it will create new entery but if key already exist then it will update the value of the key. 
dict2 = {"python" : 99, 1: "C", 2: "C++", 3: "Java", 0 : "Cobol"}
Dic_structure.update(dict2)
print(Dic_structure)

# Accessing Items from Dictionary
# Get the value of the "model" key:
thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"]
print (x)

# 7. dict.copy() method to create a duplicate copy into new memory location variable
thatdict = thisdict.copy()
print("thatdict = ", thatdict)

#8. dict.clear() - this will clear all items from a dictionary
thisdict.clear() # this works as -  in place, return value is None
print("thisdict is now clear", thisdict)
print("thatdict = ", thatdict) # still copy of it is still maintainted. 

#9. dict.pop()  - remove from the key specified in pop and not from the end. As dictionary has unordered items - no end item here
thatdict["color"] = "Red"
print(" thatdict after adding color attribute", thatdict)
# now use pop
thatdict.pop('color')
print(" thatdict after poping out color attribute", thatdict)

thatdict["color"] = "Red"
print(" thatdict after again adding color attribute", thatdict)

#10. popitem() - randomly remove any value - no key is needed. 
thatdict.popitem()
print(" thatdict after popitem ", thatdict)
```
