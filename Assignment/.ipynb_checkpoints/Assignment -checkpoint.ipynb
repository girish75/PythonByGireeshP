```python
# Assignment : Create function to find highest Even number from a list argument. 

def highesteven(list_arg):
    even_number = []
    for item in list_arg:
        if item %2 == 0: 
            even_number.append(item)
        even_number.sort()
    return (even_number[-1])

def highesteve(list_arg):
    even_number = []
    for item in list_arg:
        if item %2 == 0: 
            even_number.append(item)
    return (max(even_number))

  
def highest_even(list_arg):
    highest = 0 
    for item in list_arg:
        if item%2 == 0:
            if highest <= item: 
                highest = item
    return (highest)


print(highesteven([10,2,3,4,5,6,7,8,11]))
print(highest_even([10,2,3,4,5,6,7,8,11]))  
print(highesteve([10,2,3,4,5,6,7,8,11]))
      
```
